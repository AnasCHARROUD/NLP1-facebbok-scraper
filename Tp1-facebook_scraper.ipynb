# Stemmer, tokenization, cleaning, Remove stopwords

```python
#stopwords
#========================================================
nltk.download('stopwords')
stopwords_ar=stopwords.words('arabic')
#========================================================


stemmer =PorterStemmer()

#lire le fichier qui contient les postes
#=======================================================
myfile = open("ANASCHARROUD.csv",encoding = "UTF-8")
text = myfile.read()
#=======================================================


#netoyage de fichier
#======================================
text = re.sub('[^ء-ي]+',',',text)
#======================================


#tokenization
#=======================================
tokenizer=RegexpTokenizer(r'\w+')
tokens=tokenizer.tokenize(text)
#======================================

#eliination des stopwords
#=======================================
tokens_clean=[t for t in tokens if t not in stopwords_ar ]
#=======================================



#stemmer 
#=======================================
stems=[stemmer.stem(t) for t in tokens_clean]
#=======================================



print('__________________\n',stems)

      






```

```python
c=np.savetxt("ANASCHARROUD1.txt",stems,delimiter=',',encoding="utf-8",fmt='%s')
```
